### ЗАДАЧА: Управление складскими остатками по паттерну `MVC`

Операционная команда ведет складские остатки через простую внутреннюю панель.
Нужно реализовать систему по паттерну `MVC`, где:
- `Model` хранит товары и правила изменения остатков,
- `View` показывает текущие данные и ошибки,
- `Controller` обрабатывает действия пользователя.

НЕОБХОДИМО:

1. Реализовать `InventoryItem` как сущность модели.
2. Реализовать `InventoryModel`, который умеет:
   - добавлять товар,
   - пополнять остаток,
   - списывать остаток,
   - возвращать список товаров для отображения.
3. Бизнес-правила модели:
   - нельзя добавить товар с уже существующим `sku`,
   - нельзя пополнять или списывать несуществующий товар,
   - нельзя списать товара больше, чем есть на складе.
4. Реализовать `InventoryView`, который умеет:
   - показывать список остатков,
   - показывать сообщение об успешной операции,
   - показывать ошибку.
5. Реализовать `InventoryController`, который связывает `Model` и `View`.
6. Загрузить начальные товары из `initial_items` в модель.
7. Обработать все действия из `actions` через `controller`.
8. В конце вывести финальное состояние склада.


```python
from dataclasses import dataclass


initial_items = [
    ("SKU-001", "Laptop", 5),
    ("SKU-002", "Mouse", 20),
    ("SKU-003", "Keyboard", 12),
]

actions = [
    ("show",),
    ("restock", "SKU-002", 15),
    ("ship", "SKU-001", 2),
    ("add", "SKU-004", "Monitor", 7),
    ("ship", "SKU-003", 20),
    ("restock", "SKU-404", 5),
    ("show",),
]


@dataclass
class InventoryItem:
    sku: str
    title: str
    quantity: int


class InventoryModel:
    def __init__(self):
        # TODO 1: создайте self.items как пустой словарь
        self.items = {}
        

    def add_item(self, sku: str, title: str, quantity: int) -> None:
        # TODO 2: если sku уже существует, вызовите ValueError('Item already exists')
        if sku in self.items:
            raise ValueError('Item already exists')
        # TODO 3: создайте InventoryItem и сохраните в self.items
        self.items[sku] = InventoryItem(sku, title, quantity)
        

    def restock_item(self, sku: str, quantity: int) -> None:
        # TODO 4: если sku не найден, вызовите ValueError('Item not found')
        if sku not in self.items:
            raise ValueError('Item not found')
        # TODO 5: увеличьте quantity у товара
        self.items[sku].quantity += quantity
        
        

    def ship_item(self, sku: str, quantity: int) -> None:
        # TODO 6: если sku не найден, вызовите ValueError('Item not found')
        if sku is None:
            raise ValueError('Item not found')
        # TODO 7: если остатка меньше quantity, вызовите ValueError('Not enough stock')
        if self.items[sku].quantity < quantity:
            raise ValueError('Not enough stock')
        # TODO 8: уменьшите quantity у товара
        self.items[sku].quantity -= quantity
        

    def list_items(self) -> list[str]:
        # TODO 9: верните список строк формата:
        #   'sku | title | quantity'
        return [f'{item.sku} | {item.title} | {item.quantity}' for item in self.items.values()]


class InventoryView:
    @staticmethod
    def render_items(rows: list[str]) -> None:
        # TODO 10: если rows пустой, выведите 'Склад пуст'
        if not rows:
            print('Склад пуст')
        # TODO 11: иначе выведите заголовок 'Остатки на складе:' и все строки
        else:
            print(f'Остатки на складе:{rows}')


    @staticmethod
    def render_success(message: str) -> None:
        # TODO 12: выведите сообщение успеха
        print(message)
        

    @staticmethod
    def render_error(message: str) -> None:
        # TODO 13: выведите сообщение ошибки
        print(message)


class InventoryController:
    def __init__(self, model: InventoryModel, view: InventoryView):
        # TODO 14: сохраните model и view
        self.model = model
        self.view = view

    def add_item(self, sku: str, title: str, quantity: int) -> None:
        # TODO 15: через try/except вызовите model.add_item(...)
        try:
            model.add_item(sku,title,quantity)
            self.view.render_success('success')
        except:
            self.view.render_error('error')
        # TODO 16: покажите success или error
        

    def restock_item(self, sku: str, quantity: int) -> None:
        # TODO 17: через try/except вызовите model.restock_item(...)
        try:
            model.restock_item(sku,quantity)
            self.view.render_success('success')
        # TODO 18: покажите success или error
        except:
            self.view.render_error('error')
        

    def ship_item(self, sku: str, quantity: int) -> None:
        # TODO 19: через try/except вызовите model.ship_item(...)
        try:
            model.ship_item(sku,quantity)
            self.view.render_success('success')
        # TODO 20: покажите success или error
        except:
            self.view.render_error('error')
        

    def show_items(self) -> None:
        # TODO 21: получите строки из model.list_items() и передайте их во view.render_items(...)
        view.render_items(model.list_items())


model = InventoryModel()
view = InventoryView()
controller = InventoryController(model, view)

# TODO 22: загрузите начальные товары из initial_items в model
# Подсказка: пройдите по initial_items циклом for
# и вызовите model.add_item(sku, title, quantity)
for sku, title, quantity in initial_items:
    model.add_item(sku, title, quantity)

# TODO 23: обработайте все действия из actions
# TODO 24: если action[0] == 'show', вызовите controller.show_items()
for action in actions:
    if action[0] == 'show':
        controller.show_items()
# TODO 25: если action[0] == 'add', распакуйте sku, title, quantity
# и вызовите controller.add_item(...)
    if action[0] == 'add':
        sku, title, quantity,_ = action
        controller.add_item(sku, title, quantity)
# TODO 26: если action[0] == 'restock', распакуйте sku и quantity
# и вызовите controller.restock_item(...)
    if action[0] == 'restock':
        sku,quantity,_ = action
        controller.restock_item(sku,quantity)
# TODO 27: если action[0] == 'ship', распакуйте sku и quantity
# и вызовите controller.ship_item(...)
    if action[0] == 'ship':
        sku,quantity,_ = action
        controller.ship_item(sku,quantity)

print()
print("Финальное состояние склада:")
# TODO 28: покажите финальное состояние склада через controller.show_items()
controller.show_items()
```
