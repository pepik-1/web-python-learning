### ЗАДАЧА: Доступ к финансовому отчету через `Proxy`

Финансовая команда работает с тяжёлым дневным отчетом по транзакциям.
Проблема в том, что:
- полный отчет дорого загружать каждый раз,
- не всем сотрудникам можно его просматривать,
- нужен контроль, кто и когда запрашивал доступ.

Здесь подходит паттерн `Proxy`:
объект-заместитель будет иметь тот же интерфейс, что и реальный отчет,
но возьмет на себя:
- проверку роли пользователя,
- ленивую загрузку тяжелого отчета,
- аудит обращений.

НЕОБХОДИМО:

1. Реализовать класс `HeavyFinanceReport`.
2. В конструкторе `HeavyFinanceReport` имитировать тяжелую загрузку данных.
3. Реализовать метод `get_summary()` у реального отчета.
4. Реализовать класс `FinanceReportProxy`.
5. Внутри proxy хранить:
   - имя пользователя,
   - роль пользователя,
   - ссылку на реальный отчет,
   - журнал аудита.
6. Доступ к отчету разрешать только ролям:
   - `finance_admin`
   - `finance_analyst`
7. Если роль не подходит, proxy не должен создавать реальный отчет.
8. Если роль подходит, реальный отчет должен создаваться только при первом запросе.
9. Метод `get_summary()` у proxy должен:
   - записывать событие в аудит,
   - проверять доступ,
   - лениво создавать реальный отчет,
   - возвращать итог либо сообщение об отказе.
10. Обработать всех пользователей из `users`.
11. Вывести:
   - результаты запросов,
   - общий журнал аудита,
   - список пользователей с доступом,
   - список пользователей без доступа.


```python
users = [
    {"username": "anna", "role": "finance_admin"},
    {"username": "ivan", "role": "support_manager"},
    {"username": "olga", "role": "finance_analyst"},
    {"username": "max", "role": "marketing_manager"},
]


class HeavyFinanceReport:
    def __init__(self):
        # TODO 1: выведите строку 'Heavy report loading...'
        print('Heavy report loading...')
        # TODO 2: создайте self.data со сводкой, например:
        #   revenue, refunds, chargebacks, net_total
        self.data = {'revenue':228, 'refunds':337, 'chargebacks':69, 'net_total':52}
        

    def get_summary(self):
        # TODO 3: верните self.data
        return self.data



class FinanceReportProxy:
    allowed_roles = {"finance_admin", "finance_analyst"}

    def __init__(self, username: str, role: str):
        # TODO 4: сохраните username и role
        self.username = username
        self.role = role
        # TODO 5: создайте self._real_report = None
        self._real_report = None
        # TODO 6: создайте self.audit_log = []
        self.audit_log = []

    def _has_access(self) -> bool:
        # TODO 7: верните True, если роль пользователя входит в allowed_roles
        if self.role in self.allowed_roles:
            return True

    def _load_report_if_needed(self):
        # TODO 8: если self._real_report is None, создайте HeavyFinanceReport()
        if self._real_report is None:
            self._real_report = HeavyFinanceReport()

    def get_summary(self):
        # TODO 9: добавьте в audit_log строку:
        #   f"request:{self.username}:{self.role}"
        self.audit_log.append(f"request:{self.username}:{self.role}")

        # TODO 10: если доступа нет:
        #   - добавьте в audit_log строку f"denied:{self.username}"
        #   - верните словарь:
        #     {'status': 'denied', 'message': 'Access denied'}
        if self._has_access() != True:
            self.audit_log.append(f"denied:{self.username}")
            return {'status': 'denied', 'message': 'Access denied'}


        # TODO 11: если доступ есть:
        #   - вызовите _load_report_if_needed()
        #   - добавьте в audit_log строку f"granted:{self.username}"
        #   - верните словарь:
        #     {'status': 'ok', 'report': ...}
        if self._has_access() == True:
            self._load_report_if_needed()
            self.audit_log.append(f"granted:{self.username}")
            return {'status': 'ok', 'report': ...}


results = []
global_audit_log = []
allowed_users = []
denied_users = []

for user in users:
    proxy = FinanceReportProxy(user["username"], user["role"])

    # TODO 12: вызовите proxy.get_summary() и сохраните в result
    result = proxy.get_summary()

    results.append({
        "username": user["username"],
        "role": user["role"],
        # TODO 13: добавьте result
        'result': result
    })

    # TODO 14: добавьте все записи из proxy.audit_log в global_audit_log
    global_audit_log.append(proxy.audit_log)
    # TODO 15: если result['status'] == 'ok', добавьте username в allowed_users
    if result['status'] == 'ok':
        allowed_users.append(user['username'])
    # TODO 16: иначе добавьте username в denied_users
    else:
        denied_users.append(user['username'])

print('Результаты запросов:')
print(results)
print()

print('Общий аудит:')
print(global_audit_log)
print()

print('Пользователи с доступом:')
print(allowed_users)
print()

print('Пользователи без доступа:')
print(denied_users)
```
