### ЗАДАЧА: Аналитика заказов с экспортом в `json` и `pickle`

У интернет-магазина есть журнал заказов в виде строк.
Нужно подготовить из него две разные структуры:
- публичный аналитический отчет в `JSON`,
- полный внутренний снимок данных в `pickle`.

НЕОБХОДИМО:

1. Преобразовать строки из `rows` в список словарей `orders`.
2. Для каждого заказа привести типы:
   - `quantity` -> `int`
   - `unit_price` -> `float`
   - `is_paid` -> `bool`
3. Для каждого заказа вычислить поле `total`.
4. Построить словарь `customers_summary`, где по каждому клиенту будет:
   - `city`
   - `orders_count`
   - `paid_orders_count`
   - `total_spent`
   - `products` (список уникальных товаров)
5. Построить словарь `city_summary`, где по каждому городу будет:
   - `orders_count`
   - `paid_revenue`
6. Найти клиента с максимальной суммой покупок и сохранить его имя в `top_customer`.
7. Собрать `public_report` в таком виде:
   - `generated_for_module`: строка с названием темы
   - `customers_summary`
   - `city_summary`
   - `top_customer`
8. Сохранить `public_report` в файл `orders_report.json`.
9. Сохранить полный снимок данных в `orders_snapshot.pkl`.
10. Прочитать оба файла обратно.
11. Вывести:
   - список заказов,
   - публичный отчет,
   - данные из `JSON`,
   - данные из `pickle`,
   - итогового лучшего клиента.


```python
import json
import pickle


# row format: order_id|customer|city|product|quantity|unit_price|status|is_paid
rows = [
    "O-201|Anna|Moscow|Keyboard|2|3500.00|delivered|true",
    "O-202|Ivan|Kazan|Mouse|1|1200.00|cancelled|false",
    "O-203|Anna|Moscow|Monitor|1|18990.00|delivered|true",
    "O-204|Olga|Sochi|Laptop Stand|3|2200.50|delivered|true",
    "O-205|Max|Perm|Keyboard|1|3500.00|processing|false",
    "O-206|Olga|Sochi|Mouse|2|1200.00|delivered|true",
    "O-207|Ivan|Kazan|Webcam|1|5400.00|delivered|true",
    "O-208|Anna|Moscow|Mouse Pad|4|450.25|delivered|true",
]

orders = []
customers_summary = {}
city_summary = {}

for row in rows:
    order_id, customer, city, product, quantity_raw, unit_price_raw, status, is_paid_raw = row.split("|")

    # TODO 1: преобразуйте quantity_raw в int и сохраните в quantity
    quantity = int(quantity_raw)
    # TODO 2: преобразуйте unit_price_raw в float и сохраните в unit_price
    unit_price = float(unit_price_raw)
    # TODO 3: преобразуйте is_paid_raw в bool и сохраните в is_paid
    is_paid = True if is_paid_raw == 'true' else False
    # TODO 4: вычислите total = quantity * unit_price
    total = quantity * unit_price

    order = {
        "order_id": order_id,
        "customer": customer,
        "city": city,
        "product": product,
        # TODO 5: добавьте quantity
        'quantity':quantity,
        # TODO 6: добавьте unit_price
        'unit_price':unit_price,
        "status": status,
        # TODO 7: добавьте is_paid
        'is_paid':is_paid,
        # TODO 8: добавьте total
        'total':total
    }

    orders.append(order)

    # TODO 9: если customer еще отсутствует в customers_summary,
    if customer not in customers_summary:
        customers_summary[customer] = {
            'city':city,
            'orders_count':0,
            'paid_orders_count':0,
            'total_spent': 0.0,
            'products':[]
                             }

    # создайте для него словарь с полями:
    #   city -> city
    #   orders_count -> 0
    #   paid_orders_count -> 0
    #   total_spent -> 0.0
    #   products -> пустой список

    # TODO 10: увеличьте customers_summary[customer]['orders_count'] на 1
    customers_summary[customer]['orders_count'] += 1
    # TODO 11: если product еще не добавлен в список products,
    # добавьте его
    
    if product not in customers_summary[customer]['products']:
        customers_summary[customer]['products'].append(product)

    # TODO 12: если is_paid == True:
    #   - увеличьте paid_orders_count
    #   - прибавьте total к total_spent
    if is_paid == True:
        customers_summary[customer]['paid_orders_count'] += 1
        customers_summary[customer]['total_spent'] += total

    # TODO 13: если city еще отсутствует в city_summary,
    # создайте для него словарь:
    #   orders_count -> 0
    #   paid_revenue -> 0.0
    if city not in city_summary:
        city_summary[city] = {
            'orders_count': 0,
            "paid_revenue":0.0
        }


    # TODO 14: увеличьте city_summary[city]['orders_count'] на 1
    city_summary[city]['orders_count'] += 1

    # TODO 15: если заказ оплачен, прибавьте total к city_summary[city]['paid_revenue']
    if is_paid == True:
        total += city_summary[city]['paid_revenue']

top_customer = None
top_total = 0

for customer, summary in customers_summary.items():

    # TODO 16: сравните summary['total_spent'] с top_total
    # TODO 17: если текущая сумма больше, обновите top_customer и top_total
    if summary['total_spent'] > top_total:
        top_customer = customer
        top_total = total

round(customers_summary[customer]['total_spent'],2)
round(city_summary[city]['paid_revenue'],2)

# TODO 18: округлите все значения total_spent в customers_summary до 2 знаков
# TODO 19: округлите все значения paid_revenue в city_summary до 2 знаков

public_report = {
    "generated_for_module": "serialization_json_pickle",
    # TODO 20: добавьте customers_summary
    'customers_summary':customers_summary,
    # TODO 21: добавьте city_summary
    'city_summary':city_summary,
    # TODO 22: добавьте top_customer
    'top_customer':top_customer
}

snapshot = {
    "orders": orders,
    "customers_summary": customers_summary,
    "city_summary": city_summary,
    "top_customer": top_customer,
}

# TODO 23: сохраните public_report в orders_report.json через json.dump
#   используйте ensure_ascii=False и indent=2
with open('orders_report.json','w') as f:
    json.dump(public_report,f,ensure_ascii=False,indent=2)

# TODO 24: сохраните snapshot в orders_snapshot.pkl через pickle.dump
with open('orders_snapshot.pkl','wb') as f:
    pickle.dump(snapshot,f)

# TODO 25: прочитайте orders_report.json в loaded_report через json.load
with open('orders_report.json','r')as f:
    loaded_report = json.load(f)

# TODO 26: прочитайте orders_snapshot.pkl в loaded_snapshot через pickle.load
with open('orders_snapshot.pkl','rb')as f:
    loaded_snapshot = pickle.load(f)


print("Список заказов:")
print(orders)
print()

print("Публичный отчет:")
print(public_report)
print()

print("Данные из JSON:")
print(loaded_report)
print()

print("Данные из pickle:")
print(loaded_snapshot)
print()

print("Лучший клиент:", top_customer)
```
