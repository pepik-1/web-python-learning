### ЗАДАЧА: Экспорт данных клиентов в `json` и `pickle`

Интернет-магазин получил список клиентов в виде строк.
Нужно подготовить данные для двух разных целей:
- публичный экспорт в `JSON` для другой системы,
- внутренний бэкап в `pickle` со всеми полями.

НЕОБХОДИМО:

1. Преобразовать строки из `rows` в список словарей `customers`.
2. Для каждого клиента привести поля к правильным типам:
   - `age` -> `int`
   - `total_spent` -> `float`
   - `is_active` -> `bool`
   - `tags` -> `list[str]`
3. Собрать список `public_customers`, где останутся только поля:
   - `customer_id`
   - `name`
   - `city`
   - `is_active`
4. Сохранить `public_customers` в файл `customers_public.json`.
5. Сохранить полный список `customers` в файл `customers_backup.pkl`.
6. Прочитать оба файла обратно в программу.
7. Вывести:
   - исходный список клиентов,
   - публичный экспорт,
   - данные, прочитанные из `JSON`,
   - данные, прочитанные из `pickle`,
   - сумму `total_spent` по всем клиентам.


```python
import json
import pickle


# row format: customer_id|name|age|city|is_active|total_spent|tags
rows = [
    "C-101|Anna|24|Moscow|true|1530.50|new,vip",
    "C-102|Ivan|31|Kazan|false|820.00|sale",
    "C-103|Olga|28|Sochi|true|2310.90|vip,premium",
    "C-104|Max|35|Perm|true|640.40|regular",
]

customers = []
public_customers = []

for row in rows:
    customer_id, name, age_raw, city, is_active_raw, total_spent_raw, tags_raw = row.split("|")
    age = int(age_raw)
    total_spent = float(total_spent_raw)
    is_active = True if is_active_raw == 'true' else False
    tags = tags_raw.split(',')

    # TODO 1: преобразуйте age_raw в int и сохраните в age
    # TODO 2: преобразуйте total_spent_raw в float и сохраните в total_spent
    # TODO 3: преобразуйте is_active_raw в bool:
    #   строка 'true' должна стать True, иначе False
    # TODO 4: разделите tags_raw по ',' и сохраните в tags
    customer = {
        "customer_id": customer_id,
        "name": name,
        # TODO 5: добавьте age
        'age':age,
        "city": city,
        # TODO 6: добавьте is_active
        'is_active':is_active,
        # TODO 7: добавьте total_spent
        "total_spent": total_spent,
        'tags':tags,
        # TODO 8: добавьте tags
    }

    customers.append(customer)

    public_customer = {
        "customer_id": customer_id,
        "name": name,
        "city": city,
        # TODO 9: добавьте is_active
        'is_active':is_active
    }
    public_customers.append(public_customer)


# TODO 10: откройте файл customers_public.json в режиме записи с encoding='utf-8'
with open('customers_public.json','w',encoding='utf-8') as f:
    json.dump(public_customers,f,ensure_ascii=False,indent=2)
with open('customers_backup.pkl','wb') as f:
    pickle.dump(customers,f)

# TODO 11: сохраните туда public_customers через json.dump(..., ensure_ascii=False, indent=2)

# TODO 12: откройте файл customers_backup.pkl в бинарном режиме записи
# TODO 13: сохраните туда customers через pickle.dump

# TODO 14: откройте customers_public.json в режиме чтения
# TODO 15: загрузите данные в loaded_public через json.load
with open('customers_public.json','r',encoding='utf-8') as f:
    loaded_public = json.load(f)

# TODO 16: откройте customers_backup.pkl в бинарном режиме чтения
# TODO 17: загрузите данные в loaded_backup через pickle.load

with open('customers_backup.pkl','rb') as f:
    loaded_backup = pickle.load(f)

total_spent_sum = 0

# TODO 18: пройдите по customers и посчитайте сумму поля total_spent

total_spent_sum = sum(customer['total_spent'] for customer in customers)


print("Полные данные клиентов:")
print(customers)
print()

print("Публичный экспорт:")
print(public_customers)
print()

print("Данные из JSON:")
print(loaded_public)
print()

print("Данные из pickle:")
print(loaded_backup)
print()

print("Сумма total_spent:", round(total_spent_sum, 2))
```
